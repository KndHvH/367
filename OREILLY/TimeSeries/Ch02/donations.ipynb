```python
# week * user
# 173 * 539 = 93247
complete_index = pd.MultiIndex.from_product((set(emails_df.week),set(emails_df.user)))
complete_index
```

```python
all_emails_df = emails_df.set_index(['week','user'])
all_emails_df
```

```python
all_emails_df = all_emails_df.reindex(complete_index, fill_value=0)
all_emails_df
```

```python
all_emails_df = all_emails_df.reset_index()
all_emails_df
```

```python
all_emails_df.columns = ['week','user','n_emails']
all_emails_df
```

```python
cutoff_dates = all_emails_df.groupby('user').week.agg(['min','max'])
cutoff_dates
```

```python
all_emails_df
```

```python
for _, row in cutoff_dates.iterrows():
    user = row['user']
    min_val = row['min']
    max_val = row['max']

    # Drop rows where user matches and week is less than min_val
    drop_condition1 = (all_emails_df['user'] == user) & (all_emails_df['week'] < min_val)
    all_emails_df.drop(all_emails_df[drop_condition1].index, inplace=True)

    # Drop rows where user matches and week is greater than max_val
    drop_condition2 = (all_emails_df['user'] == user) & (all_emails_df['week'] > max_val)
    all_emails_df.drop(all_emails_df[drop_condition2].index, inplace=True)

all_emails_df
```
