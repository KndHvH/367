```python
# week * user
# 173 * 539 = 93247
complete_index = pd.MultiIndex.from_product((set(emails_df.week),set(emails_df.user)))
complete_index
```

```python
all_emails_df = emails_df.set_index(['week','user'])
all_emails_df
```

```python
all_emails_df = all_emails_df.reindex(complete_index, fill_value=0)
all_emails_df
```

```python
all_emails_df = all_emails_df.reset_index()
all_emails_df
```

```python
all_emails_df.columns = ['week','user','n_emails']
all_emails_df
```

```python
all_emails_df[all_emails_df.user == 998]
```

```python
all_emails_df[all_emails_df.user == 998].sort_values('week')
```

```python
cutoff_dates = all_emails_df.groupby('user').week.agg(['min','max']).reset_index()
cutoff_dates
```

```python
all_emails_df
```

```python
for _, row in cutoff_dates.iterrows():
    user = row['user']
    min = row['min']
    max = row['max']

    all_emails_df.drop(all_emails_df[all_emails_df.user == user][all_emails_df.week < min].index, inplace=True)
    all_emails_df.drop(all_emails_df[all_emails_df.user == user][all_emails_df.week > max].index, inplace=True)

all_emails_df
```
