# Aula 07 - Gráficos em Python (Matplotlib)

Matplotlib é uma das mais populares bibliotecas para construção de gráficos em Python. 

Em suas origens, matplotlib buscava seguir um approach MATLAB para construção de gráficos. Posteriormente, buscou-se adotar um paradigma orientado a objetos para a sua construção.

Este é um dos motivos que você vai encontrar alguns códigos que seguem o estilo MATLAB e outros que seguem a orientação a objetos. É aconselhável que você conheça ambas as formas, mas que pratique a última. 

Para compreender o matplotlib é importante entender a anatomia de um gráfico em Python.

## O estilo MATLAB

```python
import matplotlib.pyplot as plt
%matplotlib inline
import numpy as np
```

```python
x = np.linspace(0, 10, 100)

# criamos a figure
plt.figure()
# criamos o primeiro subplot 
plt.subplot(2, 1, 1) # (rows, columns, painel number)
# realizamos o plot no primeiro subplot
plt.plot(x, np.sin(x))

# criamos o segundo subplot
plt.subplot(2, 1, 2)
plt.plot(x, np.cos(x))
```

Essa abordagem é muito utilizada, principalmente pelos usuários que migraram de outras plataformas para o Python. 

Observe que ela é *stateful*, isto é, ela mantém o controle de qual é o subplot atual, sob o qual os comandos `plt.plot` são aplicados. 

## O estilo orientado a objetos

Outra forma (mais recomendada) é a orientada a objetos. Nela nós instanciamos um objeto `figure` e outro `axes` e por meio dessas referências controlamos melhor uma imagem gerada no matplotlib.

Vamos reproduzir o mesmo gráfico do slide anterior, porém sob este paradigma.

```python
fig, ax = plt.subplots(2) # indicamos que queremos 2 subplots
ax[0].plot(x, np.sin(x))
ax[1].plot(x, np.cos(x))
```

Você consegue notar as vantagens/desvantagens desse modo de produzir um gráfico?

### Ajustes finos

Fizemos um gráfico básico, mas queremos mais! Vamos ver como podemos melhorar. Vamos definir:

- O tamanho da figura
- O título da figura (e o tamanho da fonte)
- Os rótulos dos eixos X e Y
- Os ticks dos eixos X e Y
- Ativar Grid
- Fazer alguma anotação no gráfico
- Inserir uma legenda
- Adicionar linhas horizontais/verticiais

```python
# tamanho da figura
fig, ax = plt.subplots(figsize=(10, 8))
```

```python
# tamanho da figura
fig, ax = plt.subplots(figsize=(10, 8))

# título do gráfico
ax.set_title("Gráfico da Idade vs. Altura (em cm)",
            fontsize=20, fontstyle="italic")
```

```python
# tamanho da figura
fig, ax = plt.subplots(figsize=(10, 8))

# título do gráfico
ax.set_title("Gráfico da Idade vs. Altura (em cm)",
            fontsize=20, fontstyle="italic")

# alterar rótulos X e Y
ax.set_xlabel("Idade (anos)", fontsize=16)
ax.set_ylabel("Altura (cm)", fontsize=16)
```

```python
# tamanho da figura
fig, ax = plt.subplots(figsize=(10, 8))

# título do gráfico
ax.set_title("Gráfico da Idade vs. Altura (em cm)",
            fontsize=20, fontstyle="italic")

# alterar rótulos X e Y
ax.set_xlabel("Idade (anos)", fontsize=16)
ax.set_ylabel("Altura (cm)", fontsize=16)

# ativar o grid
ax.grid(True)
```

Você compreende que a cor preta é indicada por `k`?

Outras cores: 

b: blue

g: green

r: red

c: cyan

m: magenta

y: yellow

k: black

w: white


```python
# tamanho da figura
fig, ax = plt.subplots(figsize=(10, 8))

# título do gráfico
ax.set_title("Gráfico da Idade vs. Altura (em cm)",
            fontsize=20, fontstyle="italic")

# alterar rótulos X e Y
ax.set_xlabel("Idade (anos)", fontsize=16)
ax.set_ylabel("Altura (cm)", fontsize=16)

# ativar o grid
ax.grid(True)

# alterar os ticks dos eixos X e Y
ax.set_xticks([i for i in range(0, 60, 5)])
ax.tick_params(axis="x", labelsize=12)
ax.set_ylim(100, 200)
ax.tick_params(axis="y", labelsize=12)

# plotar os dados 
ax.scatter(x=age, y=height, c="orange", s=150, edgecolors="k")

# adicionar notações na imagem
ax.text(x=15, y=105, s="Há crescimento até 20 anos,\ndepois há uma certa estabilidade",
       fontsize=12, linespacing=2, rotation=30)
ax.text(x=22, y=185, s="Ninguém possui mais que 180 cm",
       fontsize=12)
```

```python
# tamanho da figura
fig, ax = plt.subplots(figsize=(10, 8))

# título do gráfico
ax.set_title("Gráfico da Idade vs. Altura (em cm)",
            fontsize=20, fontstyle="italic")

# alterar rótulos X e Y
ax.set_xlabel("Idade (anos)", fontsize=16)
ax.set_ylabel("Altura (cm)", fontsize=16)

# ativar o grid
ax.grid(True)

# alterar os ticks dos eixos X e Y
ax.set_xticks([i for i in range(0, 60, 5)])
ax.tick_params(axis="x", labelsize=12)
ax.set_ylim(100, 200)
ax.tick_params(axis="y", labelsize=12)

# plotar os dados 
ax.scatter(x=age, y=height, c="orange", s=150, edgecolors="k")

# adicionar notações na imagem
ax.text(x=15, y=105, s="Há crescimento até 20 anos,\ndepois há uma certa estabilidade",
       fontsize=12, linespacing=2, rotation=30)
ax.text(x=22, y=185, s="Ninguém possui mais que 180 cm",
       fontsize=12)

# adicionar as linhas 
ax.vlines(x=20, ymin=100, ymax=180, color="blue", linestyles="dashed")
ax.hlines(y=180, xmin=0, xmax=55, color="red", linestyles="dashed")
```

```python
# tamanho da figura
fig, ax = plt.subplots(figsize=(10, 8))

# título do gráfico
ax.set_title("Gráfico da Idade vs. Altura (em cm)",
            fontsize=20, fontstyle="italic")

# alterar rótulos X e Y
ax.set_xlabel("Idade (anos)", fontsize=16)
ax.set_ylabel("Altura (cm)", fontsize=16)

# ativar o grid
ax.grid(True)

# alterar os ticks dos eixos X e Y
ax.set_xticks([i for i in range(0, 60, 5)])
ax.tick_params(axis="x", labelsize=12)
ax.set_ylim(100, 200)
ax.tick_params(axis="y", labelsize=12)

# plotar os dados 
ax.scatter(x=age, y=height, c="orange", s=150, edgecolors="k")

# adicionar notações na imagem
ax.text(x=15, y=105, s="Há crescimento até 20 anos,\ndepois há uma certa estabilidade",
       fontsize=12, linespacing=2, rotation=30)
ax.text(x=22, y=185, s="Ninguém possui mais que 180 cm",
       fontsize=12)

# adicionar as linhas 
ax.vlines(x=20, ymin=100, ymax=180, color="blue", linestyles="dashed")
ax.hlines(y=180, xmin=0, xmax=55, color="red", linestyles="dashed")

# adicionar legendas
ax.legend(["Altura em cm"], loc="upper left", fontsize=14)
plt.show()
```

Marcadores

Há uma gama de marcadores para os gráficos. Veja alguns.

```python
xs, ys = np.mgrid[:4, 9:0:-1]
markers = [".", "+", ",", "x", "o", "D", "d", "", "8", "s", "p", "*", "|", "_", "h", "H", 0, 4, "<", "3",
           1, 5, ">", "4", 2, 6, "^", "2", 3, 7, "v", "1", "None", None, " ", ""]
descripts = ["point", "plus", "pixel", "cross", "circle", "diamond", "thin diamond", "",
             "octagon", "square", "pentagon", "star", "vertical bar", "horizontal bar", "hexagon 1", "hexagon 2",
             "tick left", "caret left", "triangle left", "tri left", "tick right", "caret right", "triangle right", "tri right",
             "tick up", "caret up", "triangle up", "tri up", "tick down", "caret down", "triangle down", "tri down",
             "Nothing", "default", "Nothing", "Nothing"]
fig, ax = plt.subplots(1, 1, figsize=(7.5, 4))
for x, y, m, d in zip(xs.T.flat, ys.T.flat, markers, descripts):
    ax.scatter(x, y, marker=m, s=100)
    ax.text(x + 0.1, y - 0.1, d, size=14)
ax.set_axis_off()
plt.show()
```

## Múltiplos Subplots

```python
fig, axes = plt.subplots(nrows=2, ncols=3, figsize=(10,8))
plt.subplots_adjust(hspace=0.3, wspace=0.2)
for i in range(2):
    for j in range(3):
        axes[i,j].text(0.5, 0.5, str((i,j)), fontsize=18, ha='center')
        axes[i,j].grid(False)       
        
# axes[0,0].scatter(...)

# axes[1,2].hbar(...)
```
