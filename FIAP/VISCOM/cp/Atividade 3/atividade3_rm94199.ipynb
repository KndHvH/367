```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import cv2

#Exibição na mesma tela do Jupyter
%matplotlib inline

plt.style.use('seaborn')
sns.set_style("whitegrid", {'axes.grid' : False})
```

## 2. Solução

Carregando a imagem original.

```python
imagem = cv2.imread("gize.png")
imagem = cv2.cvtColor(imagem, cv2.COLOR_BGR2RGB)

plt.figure(figsize=(20,10))
plt.imshow(imagem)
plt.title("Pirâmide")
```

Aplicando a máscara para realizar a subtração.

```python
# IMPLEMENTAR

img = imagem.copy()

mask = np.zeros((img.shape[0],img.shape[1]),np.int8)

points = np.array([[70,410],[280,170],[530,420]],np.int32)

cv2.fillPoly(mask,[points],(255,0,0))

plt.figure(figsize=(20,10))
plt.imshow(mask)
```

Realizando a operação de subtração.
*O seguinte [exemplo](https://docs.opencv.org/3.2.0/d0/d86/tutorial_py_image_arithmetics.html), da própria documentação do OpenCV, detalha um pouco sobre o uso das máscaras.*

```python
# IMPLEMENTAR

plt.figure(figsize=(20,10))
plt.imshow(cv2.bitwise_and(img,img,mask=mask))
```
